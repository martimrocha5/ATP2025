o número de elementos na lista que são maiores que o elemento passado como argumento.

```python
#-----Pergunta 1------
# a)
def contarOcor(c,lista):
    if lista ==[]:
        return 0
    else:
        m=lista[0]
        if c==m:
            return 1 + contarOcor(c,lista[1:])
        else:
            return contarOcor(c,lista[1:])
        
print(contarOcor(2,[1,1,2,4,5,1,2]))
```

argumento e devolve uma nova lista ordenada de forma crescente com os elementos das duas listas.

```python
#-----Pergunta 1------
# c)

def auxMergeListas(lista1,lista2,l):
    if lista1==[] and lista2==[]:
        return l
    elif lista1==[]:
        l.append(lista2[0])
        return auxMergeListas(lista1, lista2[1:], l)
    elif lista2==[]:
        l.append(lista1[0])
        return auxMergeListas(lista1[1:], lista2, l)
    else:
        if lista1[0]<=lista2[0]:
            l.append(lista1[0])
            l.append(lista2[0])
            return auxMergeListas(lista1[1:], lista2[1:], l)
        else:
            l.append(lista2[0])
            l.append(lista1[0])
            return auxMergeListas(lista1[1:],lista2[1:],l)

def mergeListas(lista1,lista2):
    l=[]
    return auxMergeListas(lista1,lista2,l)

print(mergeListas([0,2,4,5],[1,3,4,5,6]))
```

a) splitLista: defina uma função que recebe dois argumentos, uma lista de números e um número, e dá
como resultado um tuplo composto por duas listas, a primeira contendo os números divisíveis pelo número
passado como argumento, e a segunda contendo os números restantes.

```python
#-----Pergunta 2------
# a)

def splitLista(lista,n):
    divisivel=[]
    nDivi=[]
    for c in lista:
        if c%n==0:
            divisivel.append(c)
        else:
            nDivi.append(c)
            
    return (divisivel,nDivi)

print(splitLista([1,2,3,4,5],2))
```

b) mmc: defina uma função que calcula o mínimo múltiplo comum entre dois números inteiros.

```python
#-----Pergunta 2------
# b)

def mmc(n1,n2):
    i=1
    res=0
    cond=True
    while cond:
        a=n1*i
        if a%n2==0:
            res=a
            cond=False
        i+=1
    return res

print(mmc(2,3))
```

c) max2prod: defina uma função que recebe uma lista de números inteiros positivos e devolve o maior
produto que for possível calcular multiplicando os 2 maiores inteiros da lista.

```python
#-----Pergunta 2------
# c)

def max2prod(lista):
    res=1
    if len(lista)==0:
        res=0
    elif len(lista)==1:
        res=lista[0]
    elif len(lista)==2:
        res=lista[0]*lista[1]
    else:
        c=lista[0]
        for i in range(len(lista)):
            if lista[i]>c:
                c=lista[i]
        k=lista.index(c)
        nova=lista[:k]+lista[1+k:]
        m=nova[0]
        for n in range(len(nova)):
            if nova[n]>m:
                m=nova[n]
        res=c*m

    return res

print(max2prod([5,3,2,4,1]))
```

d) reduxInt: defina uma função que recebe um número inteiro positivo como argumento e vai adicionar
repetidamente os seu dígitos até obter apenas um dígito que é retornado como resultado.

```python
#-----Pergunta 2------
# d)

def reduxInt(n):
    i=0
    res=0
    c=str(n)
    if len(c)==1:
        res=n
    else:
        while len(c)>1:
            while i<len(c):
                res+=int(c[i])
                i+=1
            c=str(res)
            res=0
            i=0
            if len(c)==1:
                res=int(c)  
    return res

print(reduxInt(998))
```

d) existe_medico: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o nome
de um médico, e verifica se o médico existe nessa unidade hospitalar, devolvendo True caso exista e
False em caso contrário.

```python
#-----Pergunta 3------
# d)

def existe_medico(h, uni, m):
    cond=False
    for i in h:
        if i[0]==uni:
            for d in i[1]:
                if d[1]==m:
                    cond=True
    return cond

print(existe_medico(hospital, "Unidade de Cardiologia", "Dra. Maria"))
```

e) adicionar_medico: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o
nome de um médico, e caso este não exista na unidade, acrescenta-o e devolve o novo hospital com o
médico inserido.

```python
#-----Pergunta 3------
# e)

def adicionar_medico(h, uni, m):
    if not existe_medico(h, uni, m):
        for i in h:
            if i[0]==uni:
                i[1].append(((len(i[1])+1), m))
    return h

print(adicionar_medico(hospital, "Unidade de Cardiologia", "Dra. Paula"))
```
